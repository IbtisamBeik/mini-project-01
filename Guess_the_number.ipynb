```python
#The code with functions

import random
print("Hello, what is your name?")
my_name = input()

global number
number=random.randint(1,20)
print(my_name, "can you guess the number i am thinking of?\nYou have only 5 guesses")

def your_guess():
    number_of_guesses=0
    while number_of_guesses <5:
        print("Take a guess")
        guess=int(input())
        if guess <number:
            print("Your guess is too low")
            number_of_guesses+=1
        if guess>number:
            print("Your guess is too high")
            number_of_guesses+=1
        if guess == number:
            break
        return(guess)
    
if your_guess() ==number:
    print("Good job, you guessed the number. It is:", number)
if your_guess() !=number:
    print("Better luck next time!, The number is:", number)
```
